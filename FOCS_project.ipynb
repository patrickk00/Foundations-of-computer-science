### 6. For each pair (age, size), determine the average duration of the stay and the average cost of stay.

```python
import pandas as pd

# Load the dogs dataset into a DataFrame
df = pd.read_csv("dogs.csv")

# Group the rows by the "age" and "size" columns
grouped = df.groupby(["age", "size"])

# Calculate the average duration of stay and the average cost of stay for each group
average_duration = grouped["stay_duration"].mean()
average_cost = grouped["stay_cost"].mean()

# Combine the average duration and cost into a single DataFrame
result = pd.DataFrame({
    "average_duration": average_duration,
    "average_cost": average_cost
}).reset_index()

# Print the resulting DataFrame
print(result)
```

### 12. Find for duplicates in the dogs dataset. Two records are duplicates if they have (1) same breeds and sex, and (2) they share at least 90% of the words in the description field. Extra points if you find and implement a more refined for determining if two rows are duplicates.

```python
import numpy as np
from collections import defaultdict
import re
# Filter the DataFrame to only include rows with non-null description values
df = df[df['description'].notnull()]
def find_duplicates(df):
    # Use a nested dictionary to group the descriptions by breed and sex
    breed_sex_hash = defaultdict(dict)
    for _, row in df.iterrows():
        breed = row['breed_primary']
        sex = row['sex']
        if sex in breed_sex_hash[breed]:
            breed_sex_hash[breed][sex].append((row['id'], row['description']))
        else:
            breed_sex_hash[breed][sex] = [(row['id'], row['description'])]

    # Find all descriptions that have multiple IDs for each breed and sex
    duplicates = []
    for breed, sex_hash in breed_sex_hash.items():
        for sex, descriptions in sex_hash.items():
            if len(descriptions) > 1:
                # Use a hash table to store the descriptions and their corresponding IDs
                description_hash = defaultdict(list)
                for id, description in descriptions:
                    description_hash[description].append(id)

                # Find all descriptions that have multiple IDs within each breed and sex group
                for ids in description_hash.values():
                    if len(ids) > 1:
                        for id1 in ids:
                            for id2 in ids:
                                if id1 != id2:
                                    duplicates.append((id1, id2))

    return duplicates

def is_duplicate(description1, description2):
    # Split the descriptions using spaces and capital letters
    words1 = re.findall('[A-Z][^A-Z]*', description1)
    words2 = re.findall('[A-Z][^A-Z]*', description2)
    
    # Use the Jaccard index to compare the similarity of the descriptions
    intersection = set(words1).intersection(set(words2))
    union = set(words1).union(set(words2))
    if not union:
        return False
    jaccard_index = len(intersection) / len(union)

    # Return True if the descriptions have a high degree of similarity
    return jaccard_index >= 0.9


# Filter the DataFrame to only include rows with non-null description values
df = df[df['description'].notnull()]

# Find all duplicates in the dataset
duplicates = find_duplicates(df)

# Print the IDs of the dogs with duplicate descriptions
for id1, id2 in duplicates:
    if is_duplicate(df[df['id'] == id1]['description'].values[0], df[df['id'] == id2]['description'].values[0]):
        print(f"Dog IDs {id1} and {id2} have duplicate descriptions")
```
